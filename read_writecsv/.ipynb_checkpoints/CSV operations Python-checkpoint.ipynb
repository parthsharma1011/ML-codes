```python
#Dataframe concatenaion
import pandas as pd

#define a dictionary
data1 = {'name':['Parth','Shahid','sushma','raj'],
        'age':['24','23','20','34'],
        'Address':['bangalore','jaipur','punjab','goa'],
        'likes':['fb','tw','ig','ln']}

#define a dictionary
data2 = {'name':['bhootesh','praveen','anisha','rani'],
        'age':['22','27','29','35'],
        'Address':['bihar','odisa','gujrat','mumbai'],
        'likes':['vu','fb','ln','ig']}

df1 = pd.DataFrame(data1,index=[0,1,2,3])
df2 = pd.DataFrame(data2,index=[4,5,6,7])

print(df1,"\n\n",df1)

#or this
#df1 
#df2
```

```python
frames = [df1,df2]
result = pd.concat(frames)
result
```

```python
#concat with axis

#Dataframe concatenaion
import pandas as pd

#define a dictionary
data1 = {'name':['Parth','Shahid','sushma','raj'],
        'age':['24','23','20','34'],
        'Address':['bangalore','jaipur','punjab','goa'],
        'likes':['fb','tw','ig','ln']}

#define a dictionary
data2 = {'name':['bhootesh','praveen','anisha','rani'],
        'age':['22','27','29','35'],
        'Address':['bihar','odisa','gujrat','mumbai'],
        'likes':['vu','fb','ln','ig']}

df1 = pd.DataFrame(data1,index=[0,1,2,3])
df2 = pd.DataFrame(data2,index=[4,5,6,7])

result = pd.concat([df1,df2],axis=1,sort=False)
result
```

```python
#concat with axis

#Dataframe concatenaion
import pandas as pd

#define a dictionary
data1 = {'name':['Parth','Shahid','sushma','raj'],
        'age':['24','23','20','34'],
        'Address':['bangalore','jaipur','punjab','goa'],
        'likes':['fb','tw','ig','ln']}

#define a dictionary
data2 = {'name':['bhootesh','praveen','anisha','rani'],
        'age':['22','27','29','35'],
        'Address':['bihar','odisa','gujrat','mumbai'],
        'likes':['vu','fb','ln','ig']}

df1 = pd.DataFrame(data1,index=[0,1,2,3])
df2 = pd.DataFrame(data2,index=[4,5,6,7])

result = df1.append(df2)
result
```

```python
#Dataframe combination with ignoring indexes
import pandas as pd

#define a dictionary
data1 = {'name':['Parth','Shahid','sushma','raj'],
        'age':['24','23','20','34'],
        'Address':['bangalore','jaipur','punjab','goa'],
        'likes':['fb','tw','ig','ln']}

#define a dictionary
data2 = {'name':['bhootesh','praveen','anisha','rani'],
        'age':['22','27','29','35'],
        'Address':['bihar','odisa','gujrat','mumbai'],
        'likes':['vu','fb','ln','ig']}

df1 = pd.DataFrame(data1,index=[0,1,2,3])
df2 = pd.DataFrame(data2,index=[4,4,5,6])

result = pd.concat([df1,df2],ignore_index=True)
result
```

```python
#concat with drouping keys: key that seperate data 
#Dataframe combination with ignoring indexes
import pandas as pd

#define a dictionary
data1 = {'name':['Parth','Shahid','sushma','raj'],
        'age':['24','23','20','34'],
        'Address':['bangalore','jaipur','punjab','goa'],
        'likes':['fb','tw','ig','ln']}

#define a dictionary
data2 = {'name':['bhootesh','praveen','anisha','rani'],
        'age':['22','27','29','35'],
        'Address':['bihar','odisa','gujrat','mumbai'],
        'likes':['vu','fb','ln','ig']}

df1 = pd.DataFrame(data1,index=[0,1,2,3])
df2 = pd.DataFrame(data2,index=[4,5,6,7])
frames=[df1,df2]

result = pd.concat(frames,keys=['a','b'])
result
```

```python
import pandas as pd

#define a dictionary
data1 = {'name':['Parth','Shahid','sushma','raj'],
        'age':['24','23','20','34'],
        'Address':['bangalore','jaipur','punjab','goa'],
        'likes':['fb','tw','ig','ln']}

df = pd.DataFrame(data1,index=[0,1,2,3])

ser = pd.Series([1200,2000,3444,5666],name = 'Salary')

result = pd.concat([df,ser],axis=1)
result
```

```python
import pandas as pd

#define a dictionary
data1 = {'key': ['K0', 'K1', 'K2', 'K3'],
         'Name':['Jai', 'Princi', 'Gaurav', 'Anuj'], 
        'Age':[27, 24, 22, 32],}

#define a dictionary
data2 = {'key': ['K0', 'K1', 'K2', 'K3'],
         'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'], 
        'Qualification':['Btech', 'B.A', 'Bcom', 'B.hons']}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

result = pd.merge(df1,df2,on='key')
result
```
