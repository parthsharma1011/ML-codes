```python
#Read a CSV as a dataframe
import pandas as pd
df = pd.read_csv('example.csv')
df.head()

print(type(df['Hire Date'][0]))
df
```

```python
#conidering a Row as a Dataframe index
import pandas as pd 
df = pd.read_csv('example.csv', index_col='Name')
df
```

```python
#using parse_dates to change types
import pandas as pd
df = pd.read_csv('example.csv', index_col='Name', parse_dates=['Hire Date'])
df
#to see the type
print(type(df['Hire Date'][0]))
df
```

```python
#updating key values 
import pandas as pd
df = pd.read_csv('example.csv', 
            index_col='Employee', 
            parse_dates=['Hired'], 
            header=0, 
            names=['Employee', 'Hired','Salary', 'Sick Days'])
df
```

```python
#writing into csv
import pandas as pd
df = pd.read_csv('example.csv', 
            index_col='Employee', 
            parse_dates=['Hired'], 
            header=0, 
            names=['Employee', 'Hired','Salary', 'Sick Days'])
df
df.to_csv('modifiedcsv.csv')
```
